```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.sec.gov/Archives/edgar/data/1681682/000120919121023042/doc4.xml' # Set up the url for XML pull

r = requests.get(url) # Request the XML code from the website
soup = bs(r.content, 'xml') # Save XML code in Beautiful Soup variable

issuer_name = soup.find('issuerName') # Use find to locate an issuerName tag 
print(issuer_name.get_text())
```
